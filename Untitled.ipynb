```python
def get_VIF_values(
    df: pd.DataFrame,
    predictors: list,
    min_non_missing: float = 0.50,  # keep cols with ≥50% non-missing
    min_rows: int = 30,             # need rows >> cols
    save_file: str = None
) -> pd.DataFrame:
    X = df[predictors].copy()

    # Screen: coverage & variance
    keep = []
    for c in X.columns:
        s = X[c]
        if s.notna().mean() >= min_non_missing and s.nunique(dropna=True) > 1:
            keep.append(c)
    X = X[keep]

    # Median-impute remaining missings to avoid row wipeout (safer than fillna(0))
    X = X.apply(lambda s: s.fillna(s.median()), axis=0)

    # Bail if not enough data
    if X.shape[1] <= 1 or X.shape[0] < max(min_rows, 2 * X.shape[1]):
        print("[VIF] Skipped: insufficient rows/columns after screening.")
        return pd.DataFrame({"feature": keep, "VIF": np.nan})

    Xc = add_constant(X, has_constant="add")

    rows = []
    for i, col in enumerate(Xc.columns):
        if col == "const":   # do not report VIF for the constant
            continue
        try:
            vif_val = float(variance_inflation_factor(Xc.values, i))
        except Exception:
            vif_val = np.nan
        rows.append({"feature": col, "VIF": vif_val})

    vif_df = (pd.DataFrame(rows)
                .sort_values("VIF", ascending=False)
                .reset_index(drop=True))

    print("\n=== VIF (differenced covariates) ===")
    print(vif_df.head(20).to_string(index=False))

    if save_file:
        try:
            vif_df.to_csv(save_file, index=False)
        except Exception as e:
            print(f"[VIF] Could not save file: {e}")
    return vif_df
```
