```python
# ---- 2) Set column names (EDIT THESE)
COL_COUNTRY = "Country"
COL_SECTOR  = "CCUS_sector"
COL_YEAR    = "Year"

# Dose should be OPERATIONAL EOR-linked capture capacity in MtCO2/yr
# Common candidates: "dose_mtco2_yr", "eor_ccus_capacity_mtpy", etc.
COL_DOSE    = "dose_level"

# Optional: if your file contains a final-sample flag, filter here
# df = df[df["in_final_sample"] == 1].copy()

# ---- 3) Basic hygiene
df = df.copy()
df[COL_YEAR] = df[COL_YEAR].astype(int)
df[COL_DOSE] = pd.to_numeric(df[COL_DOSE], errors="coerce").fillna(0.0)

# Unit id = country–sector
df["unit_id"] = df[COL_COUNTRY].astype(str) + " — " + df[COL_SECTOR].astype(str)

# Treated indicator from dose
df["treated"] = (df[COL_DOSE] > 0).astype(int)

# ---- 4) Adoption year (absorbing): first year with positive dose per unit
adopt_year = (
    df.loc[df["treated"] == 1]
      .groupby("unit_id")[COL_YEAR]
      .min()
      .rename("g_x")  # adoption year
)

# Merge adoption year back
df = df.merge(adopt_year, on="unit_id", how="left")

# ---- 5) Initial dose at adoption year (dose in year g)
df["is_g_year"] = (df[COL_YEAR] == df["g_x"])
initial_dose = (
    df.loc[df["is_g_year"] & (df["treated"] == 1)]
      .groupby("unit_id")[COL_DOSE]
      .first()
      .rename("initial_dose")
)

# ---- 6) Mean/max post-adoption dose (optional, but often useful in the table)
post = df[df["treated"] == 1].copy()
post["post"] = (post[COL_YEAR] >= post["g_x"]).astype(int)

mean_post_dose = (
    post.loc[post["post"] == 1]
        .groupby("unit_id")[COL_DOSE]
        .mean()
        .rename("mean_post_dose")
)

max_post_dose = (
    post.loc[post["post"] == 1]
        .groupby("unit_id")[COL_DOSE]
        .max()
        .rename("max_post_dose")
)

# ---- 7) Build treated-units dose table
treated_units = (
    df[["unit_id", COL_COUNTRY, COL_SECTOR]]
      .drop_duplicates()
      .merge(adopt_year, on="unit_id", how="inner")
      .merge(initial_dose, on="unit_id", how="left")
      .merge(mean_post_dose, on="unit_id", how="left")
      .merge(max_post_dose, on="unit_id", how="left")
)

# ---- 8) Dose bins (descriptive only): <1 small, [1,5) medium, >=5 large
# If you want 1 included in "small" instead, change right/labels accordingly.
bins = [-np.inf, 1.0, 5.0, np.inf]
labels = ["small (<1)", "medium (1–5)", "large (≥5)"]
treated_units["dose_bin"] = pd.cut(treated_units["initial_dose"], bins=bins, labels=labels, right=False)

# Sort nicely
treated_units = treated_units.sort_values(["g_x", COL_COUNTRY, COL_SECTOR]).reset_index(drop=True)

# ---- 9A) Output: treated units dose-level table (Table A.5-style)
cols = [COL_COUNTRY, COL_SECTOR, "g_x", "initial_dose", "mean_post_dose", "max_post_dose", "dose_bin"]
print("\nTREATED UNITS DOSE TABLE:")
print(treated_units[cols].to_string(index=False))

# ---- 9B) Output: bin summary table (counts/shares + dose stats)
bin_summary = (
    treated_units.groupby("dose_bin", dropna=False)
    .agg(
        n_units=("unit_id", "nunique"),
        mean_initial=("initial_dose", "mean"),
        median_initial=("initial_dose", "median"),
        min_initial=("initial_dose", "min"),
        max_initial=("initial_dose", "max"),
    )
    .reset_index()
)

bin_summary["share"] = bin_summary["n_units"] / bin_summary["n_units"].sum()

print("\nBIN SUMMARY:")
print(bin_summary.to_string(index=False))
```

```python
def format_attgt_table(df, caption, label, tau_range=None):
    from collections import defaultdict

    def format_cell(att, se):
        if pd.isna(att):
            return "--"
        return f"{att:.3f} ({se:.3f})"

    if isinstance(df.index, pd.MultiIndex):
        df = df.reset_index()

    # Check column names
    required_cols = {'cohort', 'time', 'att', 'std_error'}
    if not required_cols.issubset(df.columns):
        raise ValueError(f"DataFrame must contain columns: {required_cols}. Found: {set(df.columns)}")

    df["tau"] = df["time"] - df["cohort"]

    if tau_range:
        df = df[(df["tau"] >= tau_range[0]) & (df["tau"] <= tau_range[1])]

    table_data = defaultdict(dict)
    for _, row in df.iterrows():
        g = int(row["cohort"])
        tau = int(row["tau"])
        table_data[g][tau] = format_cell(row["att"], row["std_error"])

    cohorts = sorted(table_data.keys())
    taus = sorted({tau for values in table_data.values() for tau in values})

    # Build LaTeX
    latex = "\\begin{table}[h]\n"
    latex += "  \\centering\n"
    latex += f"  \\caption{{{caption}}}\n"
    latex += f"  \\label{{{label}}}\n"
    latex += f"  \\begin{{tabular}}{{l{'c'*len(taus)}}}\n"
    latex += "    \\toprule\n"
    latex += "    Cohort $g$ & " + " & ".join([f"$\\tau={t}$" for t in taus]) + " \\\\\n"
    latex += "    \\midrule\n"

    for g in cohorts:
        row = [f"{g}"] + [table_data[g].get(t, "--") for t in taus]
        latex += "    " + " & ".join(row) + " \\\\\n"

    latex += "    \\bottomrule\n"
    latex += "  \\end{tabular}\n"
    latex += "\\end{table}"

    return latex


latex_code = format_attgt_table(results.att_test.attgt_obj.aggregate(type_of_aggregation="group", boot_iterations=999), caption='', label='')
print(latex_code)
```

```python
def format_cohort_summary_table(df_cohorts, caption, label):
    latex = "\\begin{table}[h]\n"
    latex += "  \\centering\n"
    latex += f"  \\caption{{{caption}}}\n"
    latex += f"  \\label{{{label}}}\n"
    latex += "  \\begin{tabular}{lccccc}\n"
    latex += "    \\toprule\n"
    latex += "    Country--sector & Adoption year & Initial capacity (Mt \\ce{CO2}/yr) & Dose bin & $m_{\\text{pre}}$ & $m_{\\text{post}}$\\\\\n"
    latex += "    \\midrule\n"

    for _, row in df_cohorts.iterrows():
        latex += f"    {row['unit']} & {row['adoption_year']} & {row['capacity']} & {row['dose_bin']} & {row['m_pre']} & {row['m_post']}\\\\\n"

    latex += "    \\bottomrule\n"
    latex += "  \\end{tabular}\n"
    latex += "\\end{table}"

    return latex

df_cohorts = results.att_test.attgt_obj.aggregate(type_of_aggregation="group", boot_iterations=9990)
cohort_latex = format_cohort_summary_table(
    df_cohorts,
    caption="Treated country--sector units, adoption year and support...",
    label="tab:cohorts"
)
```

```python
import pandas as pd

def bin_dose(capacity):
    print(capacity)
    """Return dose bin label based on CO2 capacity thresholds."""
    if capacity > 1 and capacity <= 0.5:
        return "Low"
    elif capacity > 0.5 and capacity <= 5:
        return "Medium"
    else:
        return "High"

def build_cohort_table(df, event_window=4):
    """
    Given panel data with treatment and capacity, produce cohort metadata summary table.
    
    Assumes df has columns: unit_id, country, sector, year, treat, capacity
    """
    # Step 1: Identify first treated year per unit
    df_treated = df[df["treated_now"] == 1]
    first_treat = df_treated.groupby("unit_id")["Year"].min().rename("adoption_year")

    # Step 2: Merge back adoption year
    df = df.merge(first_treat, on="unit_id", how="left")

    # Step 3: Filter to adoption year rows for capacity
    initial_capacity = df[df["Year"] == df["adoption_year"]].groupby("unit_id")["log_y"].first()

    # Step 4: Dose bin
    dose_bin = initial_capacity.apply(bin_dose).rename("dose_bin")

    # Step 5: m_pre and m_post counts
    def count_pre_post(group):
        a_year = group["adoption_year"]
        pre = group[(group["Year"] < a_year) & (group["Year"] >= a_year - event_window)]
        post = group[(group["Year"] > a_year) & (group["Year"] <= a_year + event_window)]
        return pd.Series({"m_pre": len(pre), "m_post": len(post)})

    counts = df.groupby("unit_id").apply(count_pre_post)

    # Step 6: Country-sector label
    labels = df.groupby("unit_id").first().apply(
        lambda row: f"{row['Country']}–{row['CCUS_sector']}", axis=1
    ).rename("country_sector")

    # Step 7: Combine everything
    cohort_df = pd.concat([labels, first_treat, initial_capacity.rename("capacity"), dose_bin, counts], axis=1)
    cohort_df = cohort_df.reset_index(drop=True).dropna()

    return cohort_df

def generate_latex_cohort_table(df, caption, label):
    latex = "\\begin{table}[h]\n"
    latex += "  \\centering\n"
    latex += f"  \\caption{{{caption}}}\n"
    latex += f"  \\label{{{label}}}\n"
    latex += "  \\begin{tabular}{lccccc}\n"
    latex += "    \\toprule\n"
    latex += "    Country--sector & Adoption year & Initial capacity (Mt \\ce{CO2}/yr) & Dose bin & $m_{\\text{pre}}$ & $m_{\\text{post}}$\\\\\n"
    latex += "    \\midrule\n"
    for _, row in df.iterrows():
        latex += f"    {row['country_sector']} & {row['adoption_year']} & {row['capacity']:.2f} & {row['dose_bin']} & {row['m_pre']} & {row['m_post']}\\\\\n"
    latex += "    \\bottomrule\n"
    latex += "  \\end{tabular}\n"
    latex += "\\end{table}"
    return latex

cohort_df = build_cohort_table(results.data.panel.copy(), event_window=4)

latex_code = generate_latex_cohort_table(
    cohort_df,
    caption="Treated country–sector units, adoption year and support...",
    label="tab:cohorts"
)

# Save to file
with open("cohort_table.tex", "w") as f:
    f.write(latex_code)
```

```python
from pathlib import Path
import json
import numpy as np
import pandas as pd

from did_study.reporting.summary import study_printable_from_didstudy


def save_study_summary(results: dict, artifact_dir: Path, label: str = "summary") -> Path:
    """
    Convert DidStudy.run() results -> StudyPrintable and save a JSON snapshot.
    """
    sp = study_printable_from_didstudy(results)

    def _np_to_py(x):
        # Make numpy types JSON-serialisable
        if isinstance(x, (np.floating, np.integer)):
            return float(x)
        if isinstance(x, (np.bool_,)):
            return bool(x)
        return x

    def _panel_info_jsonable(panel_info: dict) -> dict:
        out = dict(panel_info or {})
        # dose_series is often a pandas Series
        ds = out.get("dose_series")
        if isinstance(ds, (pd.Series, np.ndarray, list)):
            out["dose_series"] = [float(v) for v in np.asarray(ds).tolist()]
        return {k: _np_to_py(v) for k, v in out.items()}

    payload = {
        "panel_info": _panel_info_jsonable(sp.panel_info),
        "att_pooled": {k: _np_to_py(v) for k, v in (sp.att_pooled or {}).items()},
        "att_bins": (
            sp.att_bins.to_dict(orient="records")
            if isinstance(sp.att_bins, pd.DataFrame) and not sp.att_bins.empty
            else []
        ),
        "es_pooled": (
            sp.es_pooled.to_dict(orient="records")
            if isinstance(sp.es_pooled, pd.DataFrame) and not sp.es_pooled.empty
            else []
        ),
        "pta_pooled_p": _np_to_py(sp.pta_pooled_p),
        "es_wcb_p": _np_to_py(sp.es_wcb_p),
        "honest_did": sp.honest_did,
        "support_by_tau": (
            sp.support_by_tau.to_dict(orient="records")
            if isinstance(sp.support_by_tau, pd.DataFrame) and not sp.support_by_tau.empty
            else []
        ),
        "wcb_bins_allzero_p": _np_to_py(sp.wcb_bins_allzero_p),
        "wcb_bins_equal_p": _np_to_py(sp.wcb_bins_equal_p),
        "wcb_bins_info": sp.wcb_bins_info,
    }

    artifact_dir = Path(artifact_dir)
    artifact_dir.mkdir(parents=True, exist_ok=True)
    out_path = artifact_dir / f"study_printable_{label}.json"

    with out_path.open("w", encoding="utf-8") as f:
        json.dump(payload, f, indent=2, default=_np_to_py)

    return out_path
```
