```python
"""
pt_helpers.py

Utilities for working with PowerTuner results from pt.py:

a) load_all_experiments      -> DataFrame of all saved experiments
b) load_fully_robust         -> only experiments that "pass everything"
c) rerun_study_from_row      -> reconstruct and run DidStudy from one result row

Usage examples (from a Python shell):

    import pt_helpers as ph

    # a) load all experiments
    all_df = ph.load_all_experiments()

    # b) load only fully robust configs (passed everything)
    robust_df = ph.load_fully_robust(all_df)
    print(robust_df[[
        "candidate_id", "estimator", "effect", "mde", "p_main",
        "pta_p", "p_wcb", "wcb_joint_p", "honest_lo", "honest_hi"
    ]])

    # c) rerun a study from the first robust configuration
    if not robust_df.empty:
        row = robust_df.iloc[0]
        result = ph.rerun_study_from_row(
            row=row,
            data_path="ds_no_prep.csv",   # adjust to your data file
            artifact_dir="_artifacts/rerun_from_results",
            use_wcb=True,
            honestdid=True,
        )
        print(result)
"""

from __future__ import annotations

import ast
from pathlib import Path
from typing import Any, Dict, Iterable, List, Optional, Tuple

import numpy as np
import pandas as pd

import pt  # your PowerTuner (pt.py)
from did_study.helpers.config import StudyConfig
from did_study.study import DidStudy
from did_study.reporting.summary import print_and_plot_summary, study_printable_from_didstudy

# ---------------------------------------------------------------------
# Helpers for parsing values back from CSV into config kwargs
# ---------------------------------------------------------------------


def _parse_covariates(val: Any) -> List[str]:
    """
    Turn whatever is in the 'covariates' column into a list of strings.

    Handles:
      - empty / NaN -> []
      - comma-separated "a,b,c" -> ["a","b","c"]
      - repr(list) like "['a', 'b']" -> ["a","b"]
      - already-a-list -> same list
    """
    if val is None or (isinstance(val, float) and np.isnan(val)):
        return []
    if isinstance(val, (list, tuple, set)):
        return [str(x) for x in val]

    if isinstance(val, str):
        s = val.strip()
        if not s:
            return []
        # Try literal_eval for strings like "['a', 'b']"
        if (s.startswith("[") and s.endswith("]")) or (s.startswith("(") and s.endswith(")")):
            try:
                parsed = ast.literal_eval(s)
                if isinstance(parsed, (list, tuple, set)):
                    return [str(x) for x in parsed]
            except Exception:
                # fall back to comma split
                pass
        # Fallback: comma-separated
        return [t.strip() for t in s.split(",") if t.strip()]

    # Fallback: single value
    return [str(val)]


def _parse_list_of_numbers_or_none(val: Any) -> Optional[List[float]]:
    """
    Parse 'dose_bins', 'dose_quantiles' style columns.

    Returns:
      - None for NaN / empty / NONE / None
      - List[float] for something that looks like [0.1, 0.5, inf]
    """
    if val is None or (isinstance(val, float) and np.isnan(val)):
        return None

    if isinstance(val, (list, tuple)):
        return [float(x) for x in val]

    if isinstance(val, str):
        s = val.strip()
        if not s:
            return None
        if s.upper() in {"NONE", "NAN", "NULL"}:
            return None
        # Remove brackets if present
        if s.startswith("[") and s.endswith("]"):
            s_inner = s[1:-1]
        else:
            s_inner = s
        parts = [p.strip() for p in s_inner.split(",") if p.strip()]
        out: List[float] = []
        for p in parts:
            pl = p.lower()
            if pl in {"inf", "+inf", "infinity", "+infinity"}:
                out.append(float("inf"))
            else:
                try:
                    out.append(float(p))
                except ValueError:
                    # If it is truly non-numeric, keep as-is (rare)
                    # but cast to float('nan') to avoid breaking StudyConfig
                    out.append(float("nan"))
        return out

    # Fallback: cannot parse cleanly
    return None


def _to_int_or_none(val: Any) -> Optional[int]:
    if val is None or (isinstance(val, float) and np.isnan(val)):
        return None
    try:
        return int(val)
    except Exception:
        return None


def _to_float_or_none(val: Any) -> Optional[float]:
    if val is None or (isinstance(val, float) and np.isnan(val)):
        return None
    try:
        return float(val)
    except Exception:
        return None


def _to_bool_or_default(val: Any, default: bool = False) -> bool:
    if isinstance(val, (bool, np.bool_)):
        return bool(val)
    if isinstance(val, (int, float)) and not np.isnan(val):
        return bool(val)
    if isinstance(val, str):
        s = val.strip().lower()
        if s in {"true", "1", "yes", "y", "on"}:
            return True
        if s in {"false", "0", "no", "n", "off"}:
            return False
    return default


# ---------------------------------------------------------------------
# a) Load all experiments from artifacts
# ---------------------------------------------------------------------


def load_all_experiments(
    artifact_dir: str | Path = "./_artifacts",
) -> pd.DataFrame:
    """
    Load all experiments that pt.py has cached into a DataFrame.

    It first tries:
        <artifact_dir>/power_tuner/power_tuner_results.csv
    and falls back to:
        <artifact_dir>/power_tuner/power_tuner_results_latest.csv
    """
    artifact_dir = Path(artifact_dir)
    pt_dir = artifact_dir / "power_tuner"

    cache_path = pt_dir / "load_all_experiments.csv"
    if not cache_path.exists():
        cache_path = pt_dir / "power_tuner_results_live.csv"
        if not cache_path.exists():
            raise FileNotFoundError(
                f"No power_tuner results CSV found in {pt_dir}. "
                "Expected power_tuner_results.csv or power_tuner_results_latest.csv"
            )

    df = pd.read_csv(cache_path)
    df = pt._normalize_estimator_labels(df)
    return df


# ---------------------------------------------------------------------
# b) Filter to experiments that “pass everything”
# ---------------------------------------------------------------------


def _add_flag_columns(df: pd.DataFrame, alpha: Optional[float] = None) -> pd.DataFrame:
    """
    Add / recompute logical flags used to define "fully robust" configs.

    Adds:
      - passes_mde     : |effect| >= mde
      - p_main         : main p-value (WCB if available, else p)
      - sig_pass       : p_main <= alpha
      - pta_support    : PTA passes via analytic or WCB
      - honestdid_pass : as stored, coerced to bool
      - wcb_p_pass     : p_wcb <= alpha (if available)
      - wcb_joint_pass : wcb_joint_p (or es_wcb_p) >= alpha  (pretrend joint test)
    """
    df = df.copy()

    if alpha is None:
        if "alpha" in df.columns and not df["alpha"].isna().all():
            alpha = float(df["alpha"].iloc[0])
        else:
            alpha = 0.05

    # passes_mde: |effect| >= mde
    if "passes_mde" not in df.columns and {"mde", "effect"} <= set(df.columns):
        df["passes_mde"] = (
            df["mde"].notna()
            & df["effect"].notna()
            & (df["effect"].abs() >= df["mde"])
        )
    elif "passes_mde" in df.columns:
        df["passes_mde"] = df["passes_mde"].fillna(False)

    # main p_main
    if "p_main" not in df.columns:
        if {"p_wcb", "p"} <= set(df.columns):
            df["p_main"] = df["p_wcb"].where(df["p_wcb"].notna(), df["p"])
        elif "p" in df.columns:
            df["p_main"] = df["p"]
        else:
            df["p_main"] = float("nan")

    # significance of main effect
    df["sig_pass"] = df["p_main"].apply(
        lambda x: isinstance(x, (int, float))
        and not np.isnan(x)
        and x <= alpha
    )

    # PTA pass flags; if missing, create
    if "pta_pass" not in df.columns:
        df["pta_pass"] = False
    if "pta_wcb_pass" not in df.columns:
        df["pta_wcb_pass"] = False
    df["pta_pass"] = df["pta_pass"].apply(pt._as_bool)
    df["pta_wcb_pass"] = df["pta_wcb_pass"].apply(pt._as_bool)

    # overall PTA support
    if "pta_support" not in df.columns:
        df["pta_support"] = False
    df["pta_support"] = (
        df["pta_support"].apply(pt._as_bool)
        | df["pta_pass"]
        | df["pta_wcb_pass"]
    )

    # HonestDiD
    if "honestdid_pass" not in df.columns:
        df["honestdid_pass"] = False
    df["honestdid_pass"] = df["honestdid_pass"].apply(pt._as_bool)

    # WCB p-value for the main estimator
    if "p_wcb" not in df.columns:
        df["p_wcb"] = float("nan")
    df["wcb_p_pass"] = df["p_wcb"].apply(
        lambda x: isinstance(x, (int, float))
        and not np.isnan(x)
        and x <= alpha
    )

    # WCB joint p-value (event-study-based). Some versions store es_wcb_p.
    if "wcb_joint_p" in df.columns:
        joint_col = "wcb_joint_p"
    elif "es_wcb_p" in df.columns:
        joint_col = "es_wcb_p"
    else:
        df["wcb_joint_p"] = float("nan")
        joint_col = "wcb_joint_p"

    df["wcb_joint_pass"] = df[joint_col].apply(
        lambda x: isinstance(x, (int, float))
        and not np.isnan(x)
        and x >= alpha  # pretrend test: we *want* high p
    )

    return df


def load_fully_robust(
    df: Optional[pd.DataFrame] = None,
    artifact_dir: str | Path = "./_artifacts",
    alpha: Optional[float] = None,
) -> pd.DataFrame:
    """
    Return a DataFrame of configurations that "pass everything", defined as:

      - |effect| >= MDE
      - p_main <= alpha  (significant main effect; WCB if available)
      - PTA pretrend passes (analytic or WCB)
      - WCB p-value <= alpha
      - WCB joint pretrend test passes (joint p >= alpha)
      - HonestDiD bands exclude zero

    If df is None, it is loaded from the artifacts directory.
    """
    if df is None:
        df = load_all_experiments(artifact_dir=artifact_dir)

    df_flags = _add_flag_columns(df, alpha=alpha)

    mask = (
        df_flags.get("passes_mde", False)
        & df_flags.get("sig_pass", False)
        & df_flags.get("pta_support", False)
        & df_flags.get("wcb_p_pass", False)
        & df_flags.get("wcb_joint_pass", False)
        & df_flags.get("honestdid_pass", False)
    )

    robust_df = df_flags.loc[mask].copy()
    return robust_df


# ---------------------------------------------------------------------
# c) Reconstruct and run a study from one saved result row
# ---------------------------------------------------------------------


def _build_config_kwargs_from_row(row: pd.Series) -> Dict[str, Any]:
    """
    Map a result row into StudyConfig kwargs (except df itself).
    Only uses columns which are present.
    """
    kwargs: Dict[str, Any] = {}

    # Booleans
    if "use_log_outcome" in row:
        kwargs["use_log_outcome"] = _to_bool_or_default(row["use_log_outcome"], False)

    # Strings
    if "supdem_mode" in row and not pd.isna(row["supdem_mode"]):
        kwargs["supdem_mode"] = str(row["supdem_mode"])
    if "outcome_mode" in row and not pd.isna(row["outcome_mode"]):
        kwargs["outcome_mode"] = str(row["outcome_mode"])

    # Integers
    if "min_pre" in row:
        val = _to_int_or_none(row["min_pre"])
        if val is not None:
            kwargs["min_pre"] = val

    if "min_post" in row:
        val = _to_int_or_none(row["min_post"])
        if val is not None:
            kwargs["min_post"] = val

    if "n_bins" in row:
        val = _to_int_or_none(row["n_bins"])
        if val is not None:
            kwargs["n_bins"] = val

    # Floats
    if "treat_threshold" in row:
        val = _to_float_or_none(row["treat_threshold"])
        if val is not None:
            kwargs["treat_threshold"] = val

    # Dose specs
    if "dose_bins" in row:
        val = row["dose_bins"]
        parsed = _parse_list_of_numbers_or_none(val)
        kwargs["dose_bins"] = parsed

    if "dose_quantiles" in row:
        val = row["dose_quantiles"]
        parsed = _parse_list_of_numbers_or_none(val)
        kwargs["dose_quantiles"] = parsed

    # Covariates
    if "covariates" in row:
        covs = _parse_covariates(row["covariates"])
        kwargs["covariates"] = covs

    return kwargs


def rerun_study_from_row(
    row: pd.Series,
    data_path: str | Path,
    artifact_dir: str | Path = "./_artifacts/rerun_from_results",
    use_wcb: bool = True,
    honestdid: bool = True,
) -> Any:
    """
    Given a row from the PowerTuner results, reconstruct a StudyConfig and
    run DidStudy with that configuration.

    Parameters
    ----------
    row : pd.Series
        A single row from the results DataFrame, e.g. from load_fully_robust().
    data_path : str or Path
        Path to the underlying panel dataset CSV used by the original runs.
    artifact_dir : str or Path
        Where to write rerun artifacts; separate from the main power_tuner dir.
    use_wcb : bool
        Whether to enable WCB in this rerun.
    honestdid : bool
        Whether to enable HonestDiD checks in this rerun.

    Returns
    -------
    result : DidStudyResult
        The result of DidStudy(cfg).run()
    """
    data_path = Path(data_path)
    df = pd.read_csv(data_path)

    artifact_dir = Path(artifact_dir)
    artifact_dir.mkdir(parents=True, exist_ok=True)

    cfg_kwargs = {
        "df": df,
        "artifact_dir": str(artifact_dir),
        "use_wcb": bool(use_wcb),
        "honestdid_enable": bool(honestdid),
    }

    cfg_kwargs.update(_build_config_kwargs_from_row(row))

    cfg = StudyConfig(**cfg_kwargs)
    study = DidStudy(cfg)

    print("[pt_helpers] Re-running config with kwargs:")
    for k, v in cfg_kwargs.items():
        if k == "df":
            continue
        print(f"  {k} = {v}")

    result = study.run()
    return result


# ---------------------------------------------------------------------
# Optional: small CLI demo
# ---------------------------------------------------------------------


def _demo():
    """
    Minimal demo when running this file directly:

        python pt_helpers.py

    Adjust data_path if needed.
    """
    print("[pt_helpers] Loading all experiments...")
    all_df = load_all_experiments()
    print(f"[pt_helpers] Loaded {len(all_df)} rows.")

    print("[pt_helpers] Filtering fully robust configurations...")
    robust_df = load_fully_robust(all_df)
    print(f"[pt_helpers] Fully robust configs: {len(robust_df)}")

    if robust_df.empty:
        print("[pt_helpers] No fully robust configurations found.")
        return

    row = robust_df.iloc[0]
    print("[pt_helpers] Re-running first robust configuration (candidate_id="
          f"{row.get('candidate_id', 'NA')}, estimator={row.get('estimator', 'NA')}).")

    # Adjust data_path to your actual dataset if different:
    result = rerun_study_from_row(
        row=row,
        data_path="ds_no_prep.csv",
        artifact_dir="_artifacts/rerun_from_results",
        use_wcb=True,
        honestdid=True,
    )
    print("[pt_helpers] Rerun completed. Result object:", type(result))
```

```python
import ast
import math
from typing import Any, Dict, Optional

import numpy as np
import pandas as pd


def config_dict_from_row(
    row: pd.Series,
    base_artifact_dir: Optional[str] = None,
) -> Dict[str, Any]:
    """
    Build a StudyConfig-style kwargs dict from a single results row.

    Parameters
    ----------
    row : pd.Series
        A row from power_tuner_results.csv / combined results DF.
    base_artifact_dir : str, optional
        If provided, sets cfg['artifact_dir'] to this path.

    Returns
    -------
    dict
        Keyword arguments you can pass to StudyConfig(df=..., **cfg).
    """
    if not isinstance(row, pd.Series):
        raise TypeError("row must be a pandas.Series (e.g. df.iloc[0])")

    cfg: Dict[str, Any] = {}

    def get(col: str) -> Any:
        if col in row.index and not pd.isna(row[col]):
            return row[col]
        return None

    # ------------------------------------------------------------------
    # Simple scalar fields
    # ------------------------------------------------------------------
    # These are exactly the fields you used in the grid / dedup logic
    scalar_cols = [
        "use_log_outcome",
        "outcome_mode",
        "supdem_mode",
        "min_pre",
        "min_post",
        "treat_threshold",
        "n_bins",
    ]

    for col in scalar_cols:
        val = get(col)
        if val is None:
            continue

        if col == "use_log_outcome":
            cfg[col] = bool(val)
        elif col in ("min_pre", "min_post", "n_bins"):
            # sometimes stored as float in CSV; coerce to int
            cfg[col] = int(val)
        elif col == "treat_threshold":
            cfg[col] = float(val)
        else:
            cfg[col] = str(val)

    # ------------------------------------------------------------------
    # dose_bins / dose_quantiles: parse into Python lists or None
    # ------------------------------------------------------------------
    def _parse_listish(v: Any):
        if v is None:
            return None
        if isinstance(v, (list, tuple, np.ndarray)):
            return list(v)

        if isinstance(v, str):
            s = v.strip()
            if not s or s.lower() in ("none", "nan"):
                return None

            # Try to literal_eval first (works for "['a','b']" and "[0.1, 0.5, inf]" if inf is quoted)
            try:
                parsed = ast.literal_eval(s)
                if isinstance(parsed, (list, tuple, np.ndarray)):
                    return list(parsed)
                return parsed
            except Exception:
                # Fallback: parse comma-separated and handle "inf"
                s2 = s.strip("[]")
                if not s2:
                    return []
                out = []
                for tok in s2.split(","):
                    t = tok.strip()
                    if not t:
                        continue
                    if t.lower() in ("inf", "np.inf", "infinity"):
                        out.append(float("inf"))
                    else:
                        out.append(float(t))
                return out

        return None

    for col in ["dose_bins", "dose_quantiles"]:
        val = get(col)
        parsed = _parse_listish(val)
        if parsed is not None:
            cfg[col] = parsed
        else:
            # Explicitly store None if column exists and was non-null-ish before
            if val is not None:
                cfg[col] = None

    # ------------------------------------------------------------------
    # Covariates: turn comma-separated or repr(list) into list[str]
    # ------------------------------------------------------------------
    cov_val = get("covariates")
    if cov_val is not None:
        cov_list: list[str] = []

        if isinstance(cov_val, str):
            s = cov_val.strip()
            if s.startswith("[") and s.endswith("]"):
                # likely a repr of a list
                try:
                    parsed = ast.literal_eval(s)
                    if isinstance(parsed, (list, tuple, np.ndarray)):
                        cov_list = [str(c).strip() for c in parsed]
                    else:
                        cov_list = [str(parsed).strip()]
                except Exception:
                    # fallback: split the inside of the brackets
                    inner = s.strip("[]")
                    cov_list = [tok.strip().strip("'").strip('"')
                                for tok in inner.split(",")
                                if tok.strip()]
            else:
                # simple comma-separated string
                cov_list = [tok.strip() for tok in s.split(",") if tok.strip()]
        elif isinstance(cov_val, (list, tuple, np.ndarray)):
            cov_list = [str(c).strip() for c in cov_val]

        cfg["covariates"] = cov_list

    # ------------------------------------------------------------------
    # Optional: artifact_dir
    # ------------------------------------------------------------------
    if base_artifact_dir is not None:
        cfg["artifact_dir"] = str(base_artifact_dir)

    return cfg
```

```python
import pandas as pd

# Define which columns define a "configuration"
CONFIG_COLS = [
    "use_log_outcome",
    "supdem_mode",
    "outcome_mode",
    "min_pre",
    "min_post",
    "treat_threshold",
    "dose_bins",
    "dose_quantiles",
    "n_bins",
    "covariates",
]


def find_duplicate_configs(df: pd.DataFrame, config_cols=None) -> pd.DataFrame:
    """
    Find and print duplicate configurations in a power-tuner results table.

    Two rows are considered the same configuration if they match on all
    `config_cols` (ignoring candidate_id, estimator, etc.).

    Returns a DataFrame containing only the duplicated configs (possibly empty).
    """
    if config_cols is None:
        config_cols = CONFIG_COLS

    # Only keep columns that actually exist in df
    config_cols_present = [c for c in config_cols if c in df.columns]
    if not config_cols_present:
        raise ValueError("None of the specified config columns are present in df.")

    # Rows that have the same config as another row (keep=False → all copies)
    dup_mask = df.duplicated(subset=config_cols_present, keep=False)
    dup_df = df.loc[dup_mask].copy()

    if dup_df.empty:
        print("[CHECK] No repeated configurations found.")
        return dup_df

    print(
        f"[CHECK] Found {len(dup_df)} rows forming "
        f"{dup_df.drop_duplicates(subset=config_cols_present).shape[0]} "
        "distinct repeated configuration(s).\n"
    )

    # Show them grouped by their config
    group_key = dup_df[config_cols_present].astype(str).agg("|".join, axis=1)
    dup_df = dup_df.assign(_config_key=group_key)

    for cfg_key, grp in dup_df.groupby("_config_key"):
        print("=" * 80)
        print("Configuration:")
        for col in config_cols_present:
            val = grp.iloc[0][col]
            print(f"  {col}: {val!r}")
        print("\nRows with this configuration:")
        cols_to_show = ["candidate_id", "estimator"] + config_cols_present
        cols_to_show = [c for c in cols_to_show if c in grp.columns]
        print(grp[cols_to_show].to_string(index=False))
        print()

    return dup_df
```

```python
import math
from pathlib import Path
from typing import Any, Dict, List, Optional

import pandas as pd

from did_study.helpers.config import StudyConfig

raw_df = pd.read_csv("ds_no_prep.csv")
# Columns in power_tuner_results.csv that encode StudyConfig fields
CONFIG_COLS = [
    "use_log_outcome",
    "supdem_mode",
    "outcome_mode",
    "min_pre",
    "min_post",
    "treat_threshold",
    "dose_bins",
    "dose_quantiles",
    "n_bins",
    "covariates",
]


def _is_nan(x: Any) -> bool:
    return isinstance(x, float) and math.isnan(x)


def _maybe_int(x: Any) -> Optional[int]:
    """Convert to int if possible, otherwise return None."""
    if x is None or _is_nan(x):
        return None
    try:
        # Handles 3, 3.0, "3"
        return int(x)
    except Exception:
        return None


def _maybe_float(x: Any) -> Optional[float]:
    """Convert to float if possible, otherwise return None."""
    if x is None or _is_nan(x):
        return None
    try:
        return float(x)
    except Exception:
        return None


def _parse_bins(value: Any) -> Optional[List[float]]:
    """
    Parse dose_bins / dose_quantiles from CSV back into list[float] or None.

    Accepts:
      - None / NaN / "NA" / "NONE" -> None
      - "[0.1, 0.5, inf]" -> [0.1, 0.5, float("inf")]
    """
    if value is None or _is_nan(value):
        return None
    if isinstance(value, (list, tuple)):
        return [float(v) for v in value]

    s = str(value).strip()
    if not s or s.upper() in {"NA", "NONE"}:
        return None

    if not (s.startswith("[") and s.endswith("]")):
        # Unexpected format – ignore / return None, or raise if you prefer
        return None

    inner = s[1:-1]
    parts = [p.strip() for p in inner.split(",") if p.strip()]
    out: List[float] = []
    for p in parts:
        pl = p.lower()
        if pl in {"inf", "+inf", "np.inf", "+np.inf"}:
            out.append(float("inf"))
        elif pl in {"-inf", "-np.inf"}:
            out.append(float("-inf"))
        else:
            out.append(float(p))
    return out


def _parse_covariates(value: Any) -> List[str]:
    """
    Parse the covariates column (stored as a comma-separated string) back
    into a list of unique covariate names.
    """
    if value is None or _is_nan(value):
        return []
    if isinstance(value, (list, tuple, set)):
        parts = [str(v).strip() for v in value]
    else:
        s = str(value).strip()
        if not s or s in {"(none)", "NA"}:
            return []
        parts = [p.strip() for p in s.split(",")]

    seen = set()
    covs: List[str] = []
    for p in parts:
        if not p:
            continue
        if p not in seen:
            seen.add(p)
            covs.append(p)
    return covs


def generate_config(
    index: int,
    df_res,
    data_df=raw_df,
    *,
    results_path: str = "./_artifacts/power_tuner/power_tuner_results.csv",
    data_path: str = "ds_no_prep.csv",
    # These mimic how pt.py builds base_kwargs / cfg_base:
    artifact_root: str = "./_artifacts",
    use_wcb: bool = True,
    honestdid_enable: bool = True,
    extra_base_kwargs: Optional[Dict[str, Any]] = None,
) -> StudyConfig:
    """
    Reconstruct a StudyConfig from a given candidate_id in power_tuner_results.csv.

    - Prefers the ATT long row for that candidate, if present.
    - Parses binning and covariate columns back into Python objects.
    - Ensures n_bins is an int (or None if truly missing).
    """
    # ------------------------------------------------------------------
    # 1. Load results and subset to this candidate
    # ------------------------------------------------------------------

    sub = df_res.iloc[index]
    if sub.empty:
        raise ValueError(f"No rows found for candidate_id={candidate_id} in {results_path}")

    # ------------------------------------------------------------------
    # 2. Load the raw data

    # ------------------------------------------------------------------
    # 3. Construct base kwargs like pt.py does for a candidate
    # ------------------------------------------------------------------
    cfg_kwargs: Dict[str, Any] = {
        "use_wcb": bool(use_wcb),
        "honestdid_enable": bool(honestdid_enable),
        "differenced": False,
        "use_lag_levels_in_diff": False
    }
    if extra_base_kwargs:
        cfg_kwargs.update(extra_base_kwargs)

    # ------------------------------------------------------------------
    # 4. Fill StudyConfig kwargs from the row’s config columns
    #     (mimicking how overrides are applied in pt.py)
    # ------------------------------------------------------------------

    # Boolean / categorical fields
    cfg_kwargs["use_log_outcome"] = bool(sub["use_log_outcome"])

    cfg_kwargs["supdem_mode"] = str(sub["supdem_mode"])

    cfg_kwargs["outcome_mode"] = str(sub["outcome_mode"])

    # Numeric trimming / threshold
    v = _maybe_int(sub["min_pre"])
    if v is not None:
        cfg_kwargs["min_pre"] = v

    v = _maybe_int(sub["min_post"])
    if v is not None:
        cfg_kwargs["min_post"] = v

    v = _maybe_float(sub["treat_threshold"])
    if v is not None:
        cfg_kwargs["treat_threshold"] = v


    nb = _maybe_int(sub["n_bins"])
    if nb is not None:
        cfg_kwargs["n_bins"] = nb

    db = _parse_bins(sub["dose_bins"])
    if db is not None:
        cfg_kwargs["dose_bins"] = db

    covs = _parse_covariates(sub["covariates"])
    cfg_kwargs["covariates"] = covs

    # Optionally prune None-valued keys (let StudyConfig defaults handle them)
    cfg_kwargs = {k: v for k, v in cfg_kwargs.items() if v is not None}

    # ------------------------------------------------------------------
    # 5. Build and return the StudyConfig
    # ------------------------------------------------------------------
    cfg = StudyConfig(df=data_df, **cfg_kwargs)
    return cfg
```

```python
#!/usr/bin/env python
"""
merge_power_tuner_live.py

Merge incremental PowerTuner results from `power_tuner_results_live.csv`
into the main `power_tuner_results.csv` cache, deduplicating configurations.

Usage (shell):
    python merge_power_tuner_live.py --pt-dir ./_artifacts/power_tuner

Inside Jupyter:
    from pathlib import Path
    from merge_power_tuner_live import merge_live_into_results
    merge_live_into_results(Path("./_artifacts/power_tuner"))
"""

import argparse
from pathlib import Path
from typing import List

import pandas as pd

ATT_LONG_LABEL = "ATT long"
ATT_DIFF_LABEL = "ATT differences"

ESTIMATOR_REMAP = {
    "ATT_long": ATT_LONG_LABEL,
    "ATT_differences": ATT_DIFF_LABEL,
    "ATT long": ATT_LONG_LABEL,
    "ATT differences": ATT_DIFF_LABEL,
}


def _normalize_estimator_labels(df: pd.DataFrame) -> pd.DataFrame:
    """Ensure estimator labels are consistent for downstream grouping/printing."""
    if "estimator" not in df.columns:
        return df
    df = df.copy()
    df["estimator"] = df["estimator"].replace(ESTIMATOR_REMAP)
    return df


def merge_live_into_results(pt_dir: Path) -> None:
    """
    Merge `power_tuner_results_live.csv` into `power_tuner_results.csv`
    under `pt_dir`, deduplicating by StudyConfig + estimator.
    """
    pt_dir = pt_dir.expanduser().resolve()
    live_path = pt_dir / "power_tuner_results_live.csv"
    results_path = pt_dir / "power_tuner_results.csv"

    print(f"[merge] PowerTuner directory: {pt_dir}")

    if not live_path.exists():
        print(f"[merge] No live file found at {live_path}; nothing to merge.")
        return

    print(f"[merge] Reading live results: {live_path}")
    live_df = pd.read_csv(live_path)
    live_df = _normalize_estimator_labels(live_df)

    if results_path.exists():
        print(f"[merge] Reading existing results: {results_path}")
        existing_df = pd.read_csv(results_path)
        existing_df = _normalize_estimator_labels(existing_df)
        n_existing = len(existing_df)
    else:
        print(f"[merge] No existing results at {results_path}; starting fresh.")
        existing_df = pd.DataFrame()
        n_existing = 0

    print(f"[merge] Live rows: {len(live_df)} | Existing rows: {n_existing}")

    # If both are empty, nothing to do
    if live_df.empty and existing_df.empty:
        print("[merge] Both live and existing results are empty; nothing to do.")
        return

    # Concatenate; we'll deduplicate below
    combined = pd.concat([existing_df, live_df], ignore_index=True)

    # Align on a consistent set of estimator labels
    combined = _normalize_estimator_labels(combined)

    # Config columns used in pt.py to define a unique StudyConfig
    config_cols: List[str] = [
        "use_log_outcome",
        "supdem_mode",
        "outcome_mode",
        "min_pre",
        "min_post",
        "treat_threshold",
        "dose_bins",
        "dose_quantiles",
        "n_bins",
        "covariates",
    ]
    config_cols_present = [c for c in config_cols if c in combined.columns]

    # Deduplicate by config + estimator, like pt.py does
    if config_cols_present:
        subset_cols: List[str] = list(config_cols_present)
        if "estimator" in combined.columns:
            subset_cols.append("estimator")

        before = len(combined)
        combined = combined.drop_duplicates(subset=subset_cols, keep="last").reset_index(
            drop=True
        )
        after = len(combined)
        print(
            f"[merge] Deduplicated configs by {subset_cols}: "
            f"{before} -> {after} rows"
        )
    else:
        print(
            "[merge] Config columns missing; skipping config-based deduplication. "
            "You may want to extend this script if needed."
        )

    # Save backup of old results (if any) before overwriting
    if results_path.exists() and n_existing > 0:
        backup_path = results_path.with_suffix(".csv.bak")
        print(f"[merge] Backing up existing results -> {backup_path}")
        results_path.replace(backup_path)

    print(f"[merge] Writing merged results -> {results_path}")
    combined.to_csv(results_path, index=False)

    print(
        f"[merge] Done. Final rows in {results_path}: {len(combined)} "
        f"(live: {len(live_df)}, existing: {n_existing})"
    )


def main(argv=None) -> None:
    parser = argparse.ArgumentParser(
        description="Merge PowerTuner live results into results cache."
    )
    parser.add_argument(
        "--pt-dir",
        type=str,
        default="./_artifacts/power_tuner",
        help="Directory containing power_tuner_results*.csv "
        "(default: %(default)s)",
    )

    # IMPORTANT: use parse_known_args so Jupyter's '-f ...' doesn't crash it
    args, _unknown = parser.parse_known_args(argv)
    merge_live_into_results(Path(args.pt_dir))


if __name__ == "__main__":
    # When run as a script, just call main() with the real sys.argv.
    # parse_known_args() will safely ignore any stray args (e.g. Jupyter's -f).
    main()
```
